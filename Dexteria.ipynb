## Active User Trends Over Time

The following plot observes the number of active users on Dexteria over the course of 90 days starting on November 23rd, 2024 and ending on February 20th, 2025.

```python
user_trends = pd.read_csv('/content/drive/MyDrive/PMC/Dexteria Cleaned Data/Dexteria Data - How Active Users are Trending.csv')
user_trends
```

```python
start_date = pd.to_datetime("2024-11-23")
user_dates = pd.DataFrame({'Nth day': range(90)})
user_dates['Date'] = start_date + pd.to_timedelta(user_dates['Nth day'], unit='D')
user_trends = user_trends.merge(user_dates)
```

## Monthly Active Users Over Time


There has been an overall increase in monthly active users.

**Late November 2024 - Mid-December 2024**
The number of monthly active users fluctuates between 55 and 60.
There are small ups and downs, but no major trend.

**Late December 2024 - Early January 2025**
Around late December to early January, MAU drops below 55, indicating a possible decline in engagement. This could be due to seasonal effects (e.g., holidays leading to reduced activity).
Strong Growth from Early

**Early January 2025**
There is a steady increase in MAU. The number of users rises sharply, surpassing 75 by mid-February. This suggests that users are returning in higher numbers or new users are being retained effectively.

**Late January to mid-February**
There are small dips, but the overall trend remains upward. This indicates positive user retention and possibly new user acquisition.


```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=user_trends, x='Date', y='30 days')
plt.xlabel("Date")
plt.ylabel("Monthly Active Users")
plt.title("Monthly Active Users from November 23rd, 2024 to February 20th, 2025")
```

# Dexteria 2 Analysis

This analysis uses data collected from Dexteria Google Analytics between December 3rd, 2024 to March 2nd, 2025 and looks at sources and devices of First-Time Downloads to gain insight of how customers are finding the app.

## First-Time Downloads by Device
With this analysis, we found that Dexteria is mostly being downloaded on:

1. Desktops (63% of downloads)
2. iPads (27.3% of downloads)
3. iPhones (9.4% of downloads)

Since there was not enough data to conduct time series analysis, we only looked at graph counts of each download by device.

```python
first_time_devices = pd.read_csv('/content/drive/MyDrive/PMC/Dexteria Cleaned Data/Dexteria 2 - First Time Download Devices.csv')

def date_to_datetime(date_str):
  date_obj = datetime.strptime(date_str, "%m/%d/%y")
  return date_obj

first_time_devices['Date'] = first_time_devices['Date'].apply(date_to_datetime)
first_time_devices = first_time_devices.rename(columns = {'Apple TV First-Time Downloads' : 'Apple TV',
                                     'Apple Vision First-Time Downloads' : 'Apple Vision',
                                     'Desktop First-Time Downloads' : 'Desktop',
                                     'iPad First-Time Downloads': 'iPad',
                                     'iPhone First-Time Downloads' : 'iPhone',
                                     'iPod First-Time Downloads' : 'iPod'})
```

```python
data = {
    "Device": ["Apple TV", "Apple Vision", "Desktop", "iPad", "iPhone", "iPod"],
    "First-Time Downloads": [
        first_time_devices['Apple TV'].sum(),
        first_time_devices['Apple Vision'].sum(),
        first_time_devices['Desktop'].sum(),
        first_time_devices['iPad'].sum(),
        first_time_devices['iPhone'].sum(),
        first_time_devices['iPod'].sum()
    ]
}

downloads_by_device = pd.DataFrame(data)
downloads_by_device = downloads_by_device.sort_values(by = 'First-Time Downloads', ascending = False)
plt.figure(figsize=(10, 6))
plot= sns.barplot(data=downloads_by_device, x='Device', y='First-Time Downloads', palette='husl')

for p in plot.patches:
    plot.annotate(
        f'{p.get_height():.1f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='bottom',
        fontsize=12,
    )

plt.xlabel("Device")
plt.ylabel("Number of Downloads")
plt.title("First-Time Downloads of Dexteria 2 by Device")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

## First-Time Downloads by Source
With this analysis, we found that Dexteria 2 is primarily learned about thorugh:

1. Purchase by an Institution (63.2%)
2. Searching "Dexteria 2" on the App Store (27.4%)
3. Browsing the App Store (8.5%)
4. App Referrer (0.9%)

```python
first_time_sources= pd.read_csv('/content/drive/MyDrive/PMC/Dexteria Cleaned Data/Dexteria 2 - First Time Download Source.csv')

def date_to_datetime(date_str):
  date_obj = datetime.strptime(date_str, "%m/%d/%y")
  return date_obj

first_time_sources['Date'] = first_time_sources['Date'].apply(date_to_datetime)
first_time_sources = first_time_sources.rename(columns = {'App Referrer First-Time Downloads' : 'App Referrer',
                                     'App Store Browse First-Time Downloads' : 'App Store Browse',
                                     'Desktop First-Time Downloads' : 'Desktop',
                                     'App Store Search First-Time Downloads': 'App Store Search',
                                     'Event Notification First-Time Downloads' : 'Event Notification',
                                     'Institutional Purchase First-Time Downloads' : 'Institutional Purchase',
                                                          'Unavailable First-Time Downloads' : 'Unavailable',
                                                          'Web Referrer First-Time Downloads' : 'Web Referrer'})

first_time_sources
```

```python
import pandas as pd
import plotly.express as px

# Replace this with your actual DataFrame or computed sums
data = {
    "Source": ["App Referrer", "App Store Browse", "App Store Search", "Event Notification", "Institutional Purchase", "Unavailable", "Web Referrer"],
    "First-Time Sources": [
        first_time_sources['App Referrer'].sum(),
        first_time_sources['App Store Browse'].sum(),
        first_time_sources['App Store Search'].sum(),
        first_time_sources['Event Notification'].sum(),
        first_time_sources['Institutional Purchase'].sum(),
        first_time_sources['Unavailable'].sum(),
        first_time_sources['Web Referrer'].sum()
    ]
}

# Convert to DataFrame and sort
downloads_by_source = pd.DataFrame(data)
downloads_by_source = downloads_by_source.sort_values(by='First-Time Sources', ascending=False)

# Create Plotly bar chart
fig = px.bar(
    downloads_by_source,
    x="Source",
    y="First-Time Sources",
    text="First-Time Sources",
    title="First-Time Downloads of Dexteria 2 by Source",
    labels={"First-Time Sources": "Number of Downloads"},
    color="First-Time Sources",
    color_continuous_scale="Blues"

)

# Customize layout and text
fig.update_traces(texttemplate='%{text:.1f}', textposition='outside')
fig.update_layout(
    xaxis_title="Source",
    width = 1250,
    yaxis_title="Number of Downloads",
    coloraxis_showscale=False,
    yaxis=dict(range=[0, downloads_by_source["First-Time Sources"].max() + 10]),
    plot_bgcolor='white',    # chart background
    paper_bgcolor='white'    # outer area around the chart
)
fig.show()
```
